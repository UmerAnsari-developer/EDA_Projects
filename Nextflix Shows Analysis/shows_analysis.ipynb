```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("netflix_titles.csv")
print(df.shape)
print(df.head())
```

```python
print(df.isnull().sum())
df.dropna(inplace = True)
print(df.shape)
```

```python
df['type'] = df['type'].dropna()
type_counts = df['type'].value_counts()
type_counts.plot(kind = 'bar', color = ['red', 'purple'][:len(type_counts)])
plt.title("Movies vs TV Shows")
plt.xlabel('Type')
plt.ylabel('Count')
plt.grid(axis = 'y')
plt.show()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'],errors='coerce')
df['year_added']= df['date_added'].dt.year
df['year_added'].value_counts().sort_index().plot(kind ='line', color =  'red')
plt.title('Content Over Time ')
plt.xlabel('Year')
plt.ylabel('Numbers of Item Added')
plt.grid(True)
plt.show()
```

```python
df['country'].value_counts().head(10).plot(kind ='bar', color ='yellow')
plt.title('Top 10 Countries with Most Content ')
plt.xlabel('Countries names')
plt.ylabel('Numbers of Item ')
plt.show()
```

```python
df['rating'].value_counts().plot(kind ='pie', autopct = '%1.1f%%')
plt.title('Rating Distribution')
plt.show()
```
